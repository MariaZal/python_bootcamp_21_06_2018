### Czarno biały filtr

```python
import matplotlib.image as mpimg
im1 = mpimg.imread('images/monkey.png')

im1[:] = (np.max(im1,axis=-1,keepdims=1)+np.min(im1,axis=-1,keepdims=1))*0.5
plt.imshow(im1)
```
